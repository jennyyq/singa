```python
import pandas as pd
import numpy as np
from scipy.signal import find_peaks
import matplotlib.pyplot as plt

# Load the data from the uploaded Excel file
file_path = '/content/数据.xlsx'
data = pd.read_excel(file_path, sheet_name='周期项')

# Convert '指标名称' column to datetime and set it as the index
data['指标名称'] = pd.to_datetime(data['指标名称'])
data.set_index('指标名称', inplace=True)

# Apply 12-month centered moving average for smoothing to numeric columns only
numeric_data = data.select_dtypes(include=[np.number])
smoothed_data = numeric_data.rolling(window=12, center=True).mean()

# Define a function to detect peaks and valleys
def detect_turning_points(series, window):
    peaks, _ = find_peaks(series, distance=window)
    valleys, _ = find_peaks(-series, distance=window)
    return peaks, valleys

# Detect initial turning points
window = 6  # N is taken as 6
turning_points = {}

for column in smoothed_data.columns:
    series = smoothed_data[column].dropna()
    peaks, valleys = detect_turning_points(series, window)
    turning_points[column] = {'peaks': peaks, 'valleys': valleys}

# Function to filter turning points based on economic significance
def filter_turning_points(peaks, valleys, min_peak_valley_distance=6, min_cycle_length=16):
    filtered_peaks = []
    filtered_valleys = []

    for peak in peaks:
        if all(abs(peak - valley) >= min_peak_valley_distance for valley in valleys):
            filtered_peaks.append(peak)

    for valley in valleys:
        if all(abs(valley - peak) >= min_peak_valley_distance for peak in peaks):
            filtered_valleys.append(valley)

    # Ensure at least min_cycle_length between consecutive peaks/valleys
    filtered_peaks = [peak for i, peak in enumerate(filtered_peaks) if i == 0 or peak - filtered_peaks[i-1] >= min_cycle_length]
    filtered_valleys = [valley for i, valley in enumerate(filtered_valleys) if i == 0 or valley - filtered_valleys[i-1] >= min_cycle_length]

    return filtered_peaks, filtered_valleys

# Filter turning points
filtered_turning_points = {}
for column, points in turning_points.items():
    peaks, valleys = points['peaks'], points['valleys']
    filtered_peaks, filtered_valleys = filter_turning_points(peaks, valleys)
    filtered_turning_points[column] = {'peaks': filtered_peaks, 'valleys': filtered_valleys}

# Function to refine turning points
def refine_turning_points(series, peaks, valleys, window):
    refined_peaks = []
    refined_valleys = []

    # Ensure each peak is the highest point within a range around the peak
    for peak in peaks:
        start = max(0, peak - window)
        end = min(len(series) - 1, peak + window)
        local_max = series.iloc[start:end+1].idxmax()
        if series[local_max] == series.iloc[start:end+1].max():
            refined_peaks.append(series.index.get_loc(local_max))

    # Ensure each valley is the lowest point within a range around the valley
    for valley in valleys:
        start = max(0, valley - window)
        end = min(len(series) - 1, valley + window)
        local_min = series.iloc[start:end+1].idxmin()
        if series[local_min] == series.iloc[start:end+1].min():
            refined_valleys.append(series.index.get_loc(local_min))

    return refined_peaks, refined_valleys

# Refine turning points iteratively
def iterative_refinement(series, initial_peaks, initial_valleys, window):
    refined_peaks, refined_valleys = refine_turning_points(series, initial_peaks, initial_valleys, window)

    iteration = 0
    while True:
        new_peaks, new_valleys = refine_turning_points(series, refined_peaks, refined_valleys, window)
        if new_peaks == refined_peaks and new_valleys == refined_valleys:
            break
        refined_peaks, refined_valleys = new_peaks, new_valleys
        iteration += 1

    return refined_peaks, refined_valleys

# Apply iterative refinement
final_turning_points = {}
for column, points in filtered_turning_points.items():
    series = data[column].dropna()
    peaks, valleys = points['peaks'], points['valleys']
    refined_peaks, refined_valleys = iterative_refinement(series, peaks, valleys, window=4)
    final_turning_points[column] = {'peaks': refined_peaks, 'valleys': refined_valleys}

# Plot the results with year as the x-axis
for column in final_turning_points:
    series = data[column].dropna()
    peaks = final_turning_points[column]['peaks']
    valleys = final_turning_points[column]['valleys']

    plt.figure(figsize=(10, 6))
    plt.plot(series, label='Original Series')
    plt.plot(smoothed_data[column], label='Smoothed Series')
    plt.scatter(series.index[peaks], series.iloc[peaks], color='red', label='Peaks')
    plt.scatter(series.index[valleys], series.iloc[valleys], color='green', label='Valleys')
    plt.title(f'Turning Points for {column}')
    plt.xlabel('Year')
    plt.xticks(series.index[::12], series.index.year[::12], rotation=45)
    plt.legend()
    plt.show()
```
